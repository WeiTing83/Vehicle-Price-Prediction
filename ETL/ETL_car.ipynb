```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

from sqlalchemy import create_engine

from config import db_password
db_string=f"postgresql://postgres:{db_password}@127.0.0.1:1037/car"
engine = create_engine(db_string)
car_df = pd.read_sql_query("select * from car.public.cars_summary", con=engine)
car_df
```

```python
car_dropna_df= car_df.dropna(how="any")
# car_dropna_df.head(20)
print(car_dropna_df.count)
```

```python
car_dropna_df.head(20)
```

```python
car_dropna_df[['manufactory','model']] =  car_dropna_df["name"].str.split(" ", n=1, expand=True)
car_dropna_df.head()
```

```python
car_clean=car_dropna_df[['manufactory',"year","selling_price","fuel","seller_type","mileage"]]
car_clean
```

```python
car_clean.dtypes
```

```python
car_over17=car_clean[car_clean["year"]>2017]

print(car_over17.count)
```

```python
car_over17
```

```python
car_over17.groupby(["manufactory"])["selling_price"].count().nlargest(7)
```

```python
contents= ['Maruti', 'Hyundai', 'Tata','Mahindra','Honda','Ford','BMW']

car_over17_top7=car_over17.loc[car_over17.apply(lambda x: x['manufactory'] in contents, axis=1)]

car_over17_top7
```

```python
import seaborn as sns
ax=sns.boxplot(x="manufactory",y="selling_price", data=car_over17_top7)

plt.title('Manufactory vs. Selling_price')

plt.show()
```

```python
df=car_over17_top7.groupby(['manufactory','year']).describe()["selling_price"]
df
```
